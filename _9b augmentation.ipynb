#      Random plate number generation (Main)

```python
for evr in range(50):
    #01 to 81 city numbers
    for i in range(9):    
        for j in range(0,10):        

            if i==8 and j==2: #81 il
                break;

            if (i!=0 or j!=0):                            
                print(str(i)+ str(j))

               #inputs: 1.param= harf_list, 2p=howmany letters, 3p=directory 4p-from 5p-to            
                rnd_dgt0,lstd0= a.rnd_list(dgt_list,1,DATA_DIR + 'plaka_default/digits', i*10, i*10 + 9)                    
                rnd_dgt1,lstd1= a.rnd_list(dgt_list,1,DATA_DIR + 'plaka_default/digits', j*10, j*10 + 9)        

                #print("dgt0 ve dgt1 ok")

                for ltr_cnt in range(1,4):

                    #middle letters
                                         #inputs: 1.param= harf_list, 2p=howmany letters, 3p=directory 4p-from 5p-to
                    rnd_ltr,lst= a.rnd_list(ltr_list,ltr_cnt,DATA_DIR + 'plaka_default/letters', 0,len(ltr_list) ) 


                    #last numbers generated varying from 2 to 4 due to middle char(s) length
                    if ltr_cnt==1: 
                        rnd_dgt2,lstd2= a.rnd_list(dgt_list,4,DATA_DIR + 'plaka_default/digits',0,len(dgt_list))                  

                    elif ltr_cnt==2:
                        rnd_dgt2,lstd2= a.rnd_list(dgt_list,3,DATA_DIR + 'plaka_default/digits',0,len(dgt_list))
                        #for 2 length letter generate first 3 numbers then 4 numbers. 34AB123 34AB1234
                        print('1&2digit+char(s)+last numbers chosen randomly:',str(rnd_dgt0) + str(rnd_dgt1) +str(rnd_ltr) + str(rnd_dgt2))

                        # tr picture added to first piece
                        lst_merge = a.add_tr(DATA_DIR, lstd0 , lstd1 , lst , lstd2)
                        #concatenate pic input: lst, dir, plate_num
                        a.img_concat( lst_merge,
                                      DATA_DIR + 'plaka_out/' , 
                                      rnd_dgt0 + rnd_dgt1+ rnd_ltr + rnd_dgt2 + '.jpg')

                        #3 ok, last 4 generate
                        rnd_ltr,lst= a.rnd_list(ltr_list,ltr_cnt,DATA_DIR + 'plaka_default/letters', 0,len(ltr_list) ) 
                        rnd_dgt2,lstd2= a.rnd_list(dgt_list,4,DATA_DIR + 'plaka_default/digits',0,len(dgt_list))                                 

                    elif ltr_cnt==3: 
                        #generate number for last 2
                        rnd_dgt2,lstd2= a.rnd_list(dgt_list,2,DATA_DIR + 'plaka_default/digits',0,len(dgt_list))                    

                    print('1&2digit+char(s)+last numbers chosen randomly:' ,str(rnd_dgt0) + str(rnd_dgt1) + str(rnd_ltr) + str(rnd_dgt2))

                    # tr picture added to first piece
                    lst_merge = a.add_tr(DATA_DIR, lstd0 , lstd1 , lst , lstd2)

                    #concatenate pic
                    a.img_concat( lst_merge,
                                  DATA_DIR + 'plaka_out/' , 
                                  rnd_dgt0 + rnd_dgt1+ rnd_ltr + rnd_dgt2 + '.jpg')

                    #'\nfile locs:',       #lstd0,'\nfile locs:', lstd1)
                    #print('3 to 5. digit chosen randomly:' ,rnd_ltr, '\nfile locs:', lst)
            
```
