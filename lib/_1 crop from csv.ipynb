```python
#csv to json, crop the plate from bigger picture
import csv
import json

data = []
with open('Out_1_201810115_01.csv') as f:
    for row in csv.DictReader(f):
        data.append(row)

json_data = json.dumps(data)
with open('Cropped.json', 'w' ) as outfile:  
        json.dump(data, outfile)
```

```python
#with open(inp_json_location + inp_json_filename) as f:
with open("Cropped.json") as f:
    source_data = json.load(f)

print ("Source data loaded. Number of rows in data:", len(source_data))
```

```python
import pandas as pd
dest_data={}
```

```python
discarded_count=0
for d in source_data:
    try:
        result = {}
        result['filename'] = d['filename']
        result['Plate'] = (d['filename']).split('#')[1]
        result['x_min']= d['pos_abs_tl_left']
        result['x_max']= d['pos_abs_tl_right']
        result['y_top']= d['pos_abs_tl_top']
        result['y_bottom']=  d['pos_abs_tl_bottom']   
        dest_data[result['filename']] = result        
        #print(result)
    except:
            discarded_count += 1
print('discarded_count:', discarded_count)
print("Number of discarded records:", discarded_count)
print("Number of converted records:", len(dest_data))
result_df = pd.DataFrame.from_dict(dest_data, orient='index')
result_df.set_index('filename', inplace=True)
print(result_df[:5])
```

```python
from skimage import io
inp_img_folder= "../data/gullseye 4608"
out_img_folder= "../data/anpr_ocr_croppedimg/"
#out_ann_folder= "../data/anpr_ocr__train/ann/"
cnt=0
for index, data in result_df.iterrows():
    try:
    #if cnt<2:
        print("filename read", inp_img_folder + "/" + index)
        img = io.imread(inp_img_folder + "/" + index)        
        #img slice and copy
        ytop,ybot,xmin,xmax = int(data['y_top']),int(data['y_bottom']),int(data['x_min']),int(data['x_max'])
        img= img[ytop:ybot , xmin:xmax , :]               
        io.imsave(out_img_folder+index,img)
    #else:
    #    break
        cnt= cnt + 1    
    except Exception as e:
        print(str(e))
        
```
