```python
import numpy as np
import pandas as pd
import os
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
path='gullseye plates'
data=[]
for filename in os.listdir(path):
    fname= filename.split('.')
    #if (len(fname[0])==9):
    #    print(fname[0])
    data.append(list(fname[0]))
```

```python
df= pd.DataFrame(data,columns=['1','2','3','4','5','6','7','8'])
```

```python
sns.countplot(data=df,x='1')
```
